```python
#audio lib 
import librosa
```

```python
# We'll need numpy for some mathematical operations
import numpy as np

# matplotlib for displaying the output
import matplotlib.pyplot as plt
%matplotlib inline

# and IPython.display for audio output
import IPython.display as ipd


# And the display module for visualization
import librosa.display
```

```python
#sesin yolu notebooka yüklenir 
audio_path = (r"D:\female\arctic_a0001.wav")
x, sr = librosa.load(audio_path)
```

```python
x.shape
```

```python
#(sr)Örnekleme hızı, Hz veya kHz cinsinden ölçülen, saniyede taşınan ses örneği sayısıdır.
sr
```

```python
#sesi oynatmaya yarar 
ipd.Audio(audio_path)
```

```python
#dalga formunun amplitüd zarfının grafiğine sahibiz.
#spectrogram
import librosa.display
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)
```

```python
#spectrogram
# 2 boyutlu dizilerde, birinci eksen zaman, ikinci eksen ise zamandır.
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
##Dikey eksen frekansları (0 ila 10kHz) gösterir ve yatay eksen klibin süresini gösterir. Tüm eylemin spektrumun altında gerçekleştiğini gördüğümüz için, frekans eksenini logaritmik bir harekete dönüştürebiliriz.
#spectrogram
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
```

```python
#MFCC
#bir spektral zarfın genel şeklini kesin bir şekilde tarif eden küçük bir dizi özelliktir (genellikle yaklaşık 10-20). 

mfccs = librosa.feature.mfcc(y = x, sr = sr).T
print (mfccs.shape)
```
