### features (öznitelikler)

meanfreq: mean frequency (in kHz) = ortalama frekans 

sd: standard deviation of frequency = frekansın standart sapması

median: median frequency (in kHz) = ortanca frekans

Q25: first quantile (in kHz) = ilk çeyrek

Q75: third quantile (in kHz) = üçüncü çeyrek

IQR: interquantile range (in kHz) = interquantile aralığı 

skew: skewness  = çarpıklık

kurt: kurtosis  =  kurtosis

sp.ent: spectral entropy = spektral entropi
  
sfm: spectral flatness = spektral düzlük

mode: mode frequency = mod frekansı

centroid: frequency centroid  = frekans centroid

peakf: peak frequency (frequency with highest energy) = tepe frekansı (en yüksek enerjili frekans)

meanfun: average of fundamental frequency measured across acoustic signal = akustik sinyal ile ölçülen temel frekansın ortalaması

minfun: minimum fundamental frequency measured across acoustic signal =  akustik sinyalde ölçülen minimum temel frekans

maxfun: maximum fundamental frequency measured across acoustic signal = akustik sinyal ile ölçülen maksimum temel frekans

meandom: average of dominant frequency measured across acoustic signal = akustik sinyal ile ölçülen baskın frekansın ortalaması

mindom: minimum of dominant frequency measured across acoustic signal = akustik sinyalde ölçülen minimum baskın frekans

maxdom: maximum of dominant frequency measured across acoustic signal = akustik sinyalde ölçülen maksimum baskın frekans
 
dfrange: range of dominant frequency measured across acoustic signal = akustik sinyalde ölçülen dominant frekans aralığı

modindx:

modulation index. Calculated as the accumulated absolute difference between adjacent measurements of fundamental 
frequencies divided by the frequency range 

modülasyon dizini. Temel bitişik ölçümler arasındaki birikmiş mutlak fark olarak hesaplanır
frekans aralığının frekans aralığına bölünmesiyle

```python
# kütüphaneler
import pandas as pd
import numpy as np 
```

```python
# veriyi okuma işlemi 
df = pd.read_csv('voice.csv')
df.head()
```

```python
# male ve female miktarı 
df['label'].value_counts()
```

```python
# datamızın boyutu
df.shape
```

```python
#datadaki kolonlar hakkında boş değer,kolon tipi ve total veriyi döndürür
df.info()
```

```python
#rastgele 3 veri 
df.sample(3)
```

```python
#featurelar arası bağlantı 
df.corr()
```

```python
# değerleri aynı olan iki kolonumuz var datanın içinde 
df1 = pd.DataFrame({'centroid': df.centroid, 'meanfreq ': df.meanfreq})  
df1.tail()
```

```python
# aynı değerlere sahip olan colonlardan biri silindi (drop centroid)
df = df.drop(columns='centroid') 

# datadaki kolonlar
print(list(df.columns))
```

```python
# male = 1 ,female = 0 1/0 formatına dönüştürme
df.label = [0 if each == "male" else 1  for each in df.label] 
```

```python
# rastgele 3 label colonu yeni değerleri 
df.label.sample(3)
```

```python
# labels
y = df.label.values
y[500:505]
```

```python
# features
x = df.drop(["label"],axis = 1)
print(list(x.columns))
```

```python
# normalization
from sklearn import preprocessing
# Scale the features / özellikleri ölçeklendirme
scaler = preprocessing.StandardScaler()

# separate features from label / etiketten ayrı özellikler
x_scaled = scaler.fit_transform(x[x.columns.values[:]]) #features / özellikler
x_scaled[0]
```

```python
# train_test split 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split( x_scaled, y, test_size=0.2006, random_state = 42)
```

```python
print(x_train.shape,x_test.shape)
```

```python
say0 = 0 
say1 = 0
for each in y_train:
    if each == 0:
        say0 += 1 # male sayısı
    else:
        say1 +=1 # female sayısı
        
print("0 sayısı : ", say0 ," 1 sayısı : " ,say1 ,"total : ", (y_train.size))
```

```python
# logistic regresyon( clasification algorithm gerçekleştirilir ve model eğitilir)
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial').fit(x_train, y_train)
```

```python
#  model yardımı ile y değerleri tahmin edilir 
y_pred = logreg.predict(x_test)
y_pred[1]
```

```python
# logregin benim yerime tahmin  et ve  doğruluğu hesaplaması için kullandığımız fonksiyon 
logreg.score(x_test,y_test)
```

```python
# accuracy(doğruluk) fonksiyonunda tahmin edilir
import sklearn.metrics as metrics

# kullanılan metrik accuracy_score  popüler olan kategorik doğruluk metriği 
metrics.accuracy_score(y_test, y_pred)
```

```python
# f1 metriği sonuçlar aynı 
metrics.f1_score(y_test, y_pred)
```

#### Perform CV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score as accuracy
```

```python
tuned_parameters = [{'kernel': ['rbf', 'linear'], 'gamma': ['auto_deprecated', 'scale'],
                     'C': np.linspace(1e-9, 1, 5)}]
```

```python
score = 'accuracy'
clf = GridSearchCV(SVC(), tuned_parameters, cv=10, scoring=score)

clf.fit(x_train, y_train)
print(f'Traınıng accuracy: {clf.best_score_} Testing accuracy: {accuracy(y_test, clf.predict(x_test))}')
```
